```python
import env
from agent import Agent
import numpy
numpy.random.seed(2)
```

```python
env_train = env.Systra(mode="train")
env_test = env.Systra(mode="test", random_data=False)
```

```python
state_size = env_train.observation_space.shape[0]
action_size = env_train.action_space.n
```

```python
player = Agent(state_size, action_size)
```

```python
class Logger(object):
    def __init__(self):
        self.record_list = []
        self.validation_list = []
        
    def record(self, d):
        self.record_list.append(d)
        
    def val(self, d):
        self.validation_list.append(d)

from collections import deque
import random

def run(player, env, logger, mode, episode=100, batch_size=32, memory_size=1024, env_test=None):
    assert mode in ["train", "test"]
    exp = deque(maxlen=memory_size)
    for e in range(episode):
        done= False
        power_list = []
        state = env.reset()
        while not done:
            action = player.get_action(state)
            next_state, reward, done, info = env.step(action)
            exp.append((state, action, reward, next_state, done))
            state = next_state
            power_list.append(env.power)
        if len(exp) >= memory_size:
            states, actions, rewards, next_states, dones = select_batch(exp, batch_size)
            player.train_model(states, actions, rewards, next_states, dones)
        
        logger.record(power_list)
        if env_test is not None:
            p = validation(player, env_test)
            logger.val(p)
        
def validation(player, env):
    done= False
    power_list = []
    state = env.reset()
    while not done:
        action_prob = player.get_action_prob(state)
        action = numpy.argmax(action_prob)
        next_state, reward, done, info = env.step(action)
        state = next_state
        power_list.append(env.power)
    return env.power
    
        
def select_batch(exp, batch_size):
    tpls = random.sample(exp, batch_size)
    states = [t[0] for t in tpls]
    actions = [t[1] for t in tpls]
    rewards = [t[2] for t in tpls]
    next_states = [t[3] for t in tpls]
    dones = [t[4] for t in tpls]
    return _v(states), _v(actions), _v(rewards), _v(next_states), _v(dones)

def _v(arr_list):
    # explode
    arr = numpy.vstack([v for v in arr_list])
    return arr
```

```python
logger_train = Logger()
run(player, env_train, logger_train, "train", episode=5000, batch_size=512, memory_size=4096, env_test=env_test)
player.actor.save_weights("model/actor_weight.h5")
player.critic.save_weights("model/critic_weight.h5")
```
